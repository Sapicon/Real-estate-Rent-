```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import requests
from bs4 import BeautifulSoup as bs
```

```python
r = requests.get('https://nigeriapropertycentre.com/for-rent')
soup = bs(r.content)
```

```python
# def get_url(url):
#     page = requests.get(url)
#     houses = bs(page.content)
#     return houses

def get_details(url):
    property_details = {}
#     get_url(url)
    page = requests.get(url)
    house = bs(page.content)
    table = house.find('table')
    content = [item.get_text().split(':') for item in table.find_all('td')]
    property_details['Period'] = house.find('span', class_='period').get_text().replace('\xa0', "")
    property_details['Price'] = house.find_all('span', class_='price')[1].get_text().replace('\xa0', "")
    property_details['Type'] = house.find('h4', class_='content-title').get_text().replace('\xa0', "")
    property_details['Location']= house.find('p').get_text().replace('\xa0', "")
    for text in content:
        property_details[text[0]] = text[-1].replace('[', "").replace(']', "").replace('\xa0', "")
    return property_details

def get_properties(url, page_num):
    housing_dataset_list = []
    page = requests.get(url)
    house = bs(page.content)
    num_pages = int(house.find('ul', class_='pagination').find_all('li')[-2].get_text())
    for page in range(1001,1045):
        print(page)
        properties = house.find_all('div', class_='property-list')
        rel_url = [item.find('div', class_='wp-block-title').a.get('href') for item in properties]
        for link in rel_url:
            housing_dataset_list.append(get_details('https://nigeriapropertycentre.com'+link))
        page_num=+1
        url = f'https://nigeriapropertycentre.com/for-rent?page={page_num}'
        page = requests.get(url)
        house = bs(page.content)  
    housing_dataset = pd.DataFrame(housing_dataset_list)
    return housing_dataset
    
```
